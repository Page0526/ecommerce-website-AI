#### Since the data has some few errors, we will handle them 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
cd = os.getcwd()
print(os.path.dirname(cd))
data_path = os.path.join(os.path.dirname(cd), 'data')
print(data_path)
```

```python
df = pd.read_csv(os.path.join(data_path,'data.csv'))
```

```python
cake_names = pd.read_csv(os.path.join(data_path, 'cake_list.csv'), index_col=False, header=None)
```

```python
cake_names.columns = [['old','new']]
```

```python
cake_names
```

```python
for index, row in cake_names.iterrows():
    old_name = row['old']
    new_name = row['new']
    df.loc[df['Items'] == old_name, 'Items'] = new_name
```

```python
# check if replacement works?
df.loc[df['Items'] == 'The BART']
```

```python
df.to_csv(os.path.join(data_path,'data.csv'),index=False)
```
